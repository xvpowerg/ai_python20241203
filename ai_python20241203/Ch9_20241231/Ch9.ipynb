### 不固定接收數量

( ) 內接收資料若是以 * 表示代表可以引入不定數量的參數：

一個*代表以 tuple 的方式引入

兩個*代表以 dict 的方式引入

接收資料若需加入多個設定，這些設定請依照順序排列：

第一個參數設定為接收值 

第二個參數設定為初始值

第三個參數設定為不固定值，可以是 tuple 或 dict 方式

```python
def printInfo(arg1,arg2="defult",*arg3):
    print("arg1:",arg1)
    print("arg2:",arg2)
    print("arg3:",arg3)
    print("="*50)
printInfo(1)
printInfo(1,2)
printInfo(1,2,3)
printInfo(1,2,3,4)
printInfo(1,2,3,4,5,6)
```

```python
def printInfo(*arg3,arg1="defult1",arg2="defult2"):
    print("arg1:",arg1)
    print("arg2:",arg2)
    print("arg3:",arg3)
    print("="*50)
printInfo(1)
printInfo(1,2,arg2="Ken")
printInfo(1,2,3)
printInfo(1,2,3,4,arg1="Vivin")
printInfo(1,2,3,4,5)
```

```python
def calulate(*values,**args):
    cType = "+"
    if "cType" in args:
        cType = args["cType"]
    result = 0
    if cType == "*":
       result = 1
    for v in values:
        if cType == "+":
                result += v
        elif cType == "*":
                result *= v
        elif cType == "-":
                result -= v
    return result 
print(calulate(3,2,4))
print(calulate(3,2,4,cType="*"))
print(calulate(3,2,4,cType="+"))
print(calulate(3,2,4,cType="-"))
```

## 匿名函數
不需要定義函數名稱，只需要用到運算式或表達分析語法。

Python使用lambda語法定義匿名函數。

匿名函數是一個表達式或計算式，並不是一個執行流程區塊。

匿名函數出現在一般函數不允許的地方，例如list裡面或著函數呼叫參數的位置。

```python
def printInfo():
    print("Hello!!")
printInfo()
print(type(printInfo))
func = printInfo
func()
```

```python
def calulateFunc(func,*values):
    result = []
    for i in values:
        result.append(func(i))
    return result
f1 = lambda x: x + 2
print(type(f1))
result = calulateFunc(f1,1,2,3,4,5)
print(result)
```

## Map 用來做轉換

```python
list3 = [1,2,3,'',4,5,'']
msgObj = map(lambda x :x if x != '' else 0, list3)
print(list(msgObj))
```

```python
list4 = ["M","W","M","M","W"]
dirc = {"M":0,"W":1}
objMap = map(lambda x:dirc[x],list4)
print(list(objMap))
```

## Filter

```python
a = [1,2,3,4,5,6]
#使用for迴圈幫我找出偶數
for i in a:
    if i % 2 == 0:
        print(i,end = " ")
```

```python
a = [1,2,3,4,5,6]
filtetObj = filter(lambda x : x % 2 == 0,a)
print(list(filtetObj))
```

```python
list3 = [1,2,3,'',4,5,'']
filtetObj = filter(lambda x : x != '',list3)
print(list(filtetObj))
```
