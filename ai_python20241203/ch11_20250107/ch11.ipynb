```python
class Student:
    def __init__(self,name,score=0):
        self.name = name
        self.score = score
    def printInfo(self):
        print(self.name,self.score)
    def test(self):
        if self.score >= 60:
            print(self.name,"及格")
            return
        print(self.name,"不及格")     
st1 = Student("Vivin",score=25)
#print(st1.name,st1.score)
st1.printInfo()
st2 = Student(score = 75,name = "Iris")
st1.test()#Vivin 不及格
st2.printInfo()
st2.test()#Iris 及格
#print(st2.name,st2.score)
```

## 私有化

```python
class Animal:
    def __init__(self,name,age):
        self.__name = name
        self.__age = age#私有化
    def printInfo(self):
        print(self.__name,self.__age)
    def setAge(self,age):
        if age > 200 or age < 0:
            raise Exception("錯誤的年齡")
        self.__age = age
    def getAge(self):
        return self.__age;
    def setName(self,name):
        if name == None or len(name) < 1:
            raise Exception("name不可為空")
        self.__name = name
    def getName(self):
        return self.__name
a1 = Animal("Ken",35)
#a1.age =75
a1.setName(None) #希望檢察 name不可為空 或 None
#如果為空拋出錯誤 不可為空
a1.setAge(75)
a1.printInfo()

```

```python
class Employee:
    def __init__(self,name,salary):
        self.__name = name
        self.__salary = salary
    def printInfo(self):
        print(self.__name,":",self.__salary)
    def setName(self,name):
        self.__name = name
    def getName(self):
        return self.__name
    def setSalary(self,salary):
        self.__salary = salary
    def getSalary(self):
        return self.__salary
    @property     
    def name(self):
        return self.getName()
    @name.setter 
    def name(self,name):
        self.setName(name)
    @property
    def salary(self):
        return self.getSalary()
    @salary.setter    
    def salary(self,salary):
        self.__salary = salary
emp1 = Employee("Ken",50000)
emp1.printInfo()
#emp1.setName("Iris")
emp1.name = "Lucy"
emp1.salary = 9000
emp1.salary += 2000
emp1.printInfo()
```
