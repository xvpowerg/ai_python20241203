## 例外語法
try:
    保護的程式區段
    
except 例外1 :
    例外1發生時執行的程式
    
except :
    其他例外發生時執行的程式
    #未指定例外型態，捕捉所有例外物件
    
else:
    若try程式區段沒產生例外執行的程式
    
finally:
    不管有沒有發生例外都會執行的程式

```python
num1 = 10
num2 = 0
nums = [1,2,3,4]
print("Start")
try:
    #print(num1/num2)
    print(nums[5])
except:
    print("有錯誤發生")
print("End")    
```

```python
num1 = 10
num2 = 0
nums = [1,2,3,4]
print("Start")
#ZeroDivisionError
#IndexError
try:
    #print(num1/num2)
    #print(nums[5])
    int("A")
except ZeroDivisionError as ex:
    print(f"Error分母不可為0:{ex}")
except IndexError:
    print("Index發生問題")
except Exception as ex:
    print(f"錯誤發生:{ex}")
```

```python
num1 = 10
num2 = 0
nums = [1,2,3,4]
print("Start")
#ZeroDivisionError
#IndexError
try:
    #print(num1/num2)
    #print(nums[5])
    int("12")
except ZeroDivisionError as ex:
    print(f"Error分母不可為0:{ex}")
except IndexError:
    print("Index發生問題")
except Exception as ex:
    print(f"錯誤發生:{ex}")
else:
    print("沒有發生任何錯誤")
finally:
    print("一定會執行一次")
```

```python
def testScore(score):
    if score > 100 or score < 0:
        print("錯誤的成績")
        return #離開函數
   
    print(f"成績:{score}")
        

testScore(50)#成績50
testScore(120)#錯誤的成績   
# score <= 100 score >= 0 如果不在這個範圍 顯示錯誤的成績    
```

## 使用raise拋例外

```python
def testScore(score):
    if score > 100 or score < 0:
        raise OverflowError("錯誤的成績")   
   
    print(f"成績:{score}")
        

testScore(50)#成績50
testScore(120)#錯誤的成績   
```

## 正規表示式




```python
import re
s = "Tim's phone numbers are 12345-41521 and 78963-85214"
match = re.findall(r"\d",s)
print(match)
```

```python
import re
s = "Tim's phone numbers are 12345-41521 and 78963-85214"
match = re.findall(r"\d{5}",s)
print(match)
```

```python
import  re
s="Python was conceived in the late 1980s by Guido van Rossum"
match = re.findall(r"[a-t]",s)
print(match)
match = re.findall(r"[a-t]{2}",s)
print(match)
match = re.findall(r"[a-t]{3,6}",s)
print(match)
match = re.findall(r"[a-t]{4,}",s)
print(match)
```

```python
import re
pattern = re.compile(r"\d+")
result1 = pattern.findall("1hello 123456 789")
result2 = pattern.findall("one1two2three3four4")
print(result1)
print(result2)
result3 = pattern.findall("one1two2three3four4",0,13)#0~不包含13
print(result3)
```

## split 方法
+ split 方法能以比對規則針對字串進行切割後產生 list。
+ split 它的使用形式如下：
+ split(字串,最大分割次數)
   + 若不指定最大分割次數則全部分割

```python
import re
text = "Ken,Vivin,Joy"
re1 = re.split(r",",text)
print(re1)
```

```python
text = 'one, two...ten'
re2 = re.split(r"[,. ]",text)
print(re2)
re3 = re.split(r"[,. ]+",text)
print(re3)
```

## sub 方法
+ sub 方法用於替換，它的使用形式如下
  + sub(取代的字串或函數,來源字串,可設定被取代的數量)
+ 可設定被取代的數量若不指定時全部替換。

```python
import re
inputStr = "hello 123 world 456"
rStr = re.sub(r"\d+","222",inputStr)
print(rStr)
rStr = re.sub(r"\d+","222",inputStr,count=1)
print(rStr)
```
