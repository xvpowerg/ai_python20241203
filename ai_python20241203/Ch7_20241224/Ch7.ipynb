## 作業 文字接龍

```python
txt = input("請輸入第一個字串")
strList = []
strList.append(txt)
msg = txt
count = 0
inputMsg = "請輸入-"+txt[-1]+"-開始的字串"
while count < 5:
    tmpTxt = input(inputMsg)
    if tmpTxt[0] == txt[-1]:
        txt = tmpTxt
        strList.append(tmpTxt)
        msg = "-".join(strList)
        inputMsg = "請輸入-"+txt[-1]+"-開始的字串"
    else:
        inputMsg = "請再次輸入-"+txt[-1]+"-開始的字串"
        count += 1
    print("上一個字串是:"+msg)    
```

## list 項目擴展資料
+ extend 擴展資料：
  + 如果引號括起來的字串有 [ ] 將會拆解成好幾個資料。
  + 可以把其他 list 加入到這個 list 內，擴展多個欄位。

```python
list1 = [1,2,3,4,5]
list2 = [22,33,44]
list1.append(list2)
print(list1)
```

```python
list1 = [1,2,3,4,5]
list2 = [22,33,44]
list1.extend(list2)
print(list1)
```

## list 項目插入新資料
+ list 可使用 insert 方式插入新的資料：
   + list.insert(索引值，插入的資料)。

### pop

如果 pop( ) 內沒有參數，則移除最後一筆資料。

pop() 內若有參數，則移除指定位置的資料。

```python
list2 = []
list2.append("Ken")
list2.append("Vivin")
list2.append("Lindy")
list2.append("Lucy")
print(list2)
v1 = list2.pop()
print(list2)
print("v1:",v1)
v2 = list2.pop()
print(list2)
print("v2:",v2)
v3 = list2.pop()
print(list2)
print("v3:",v3)
v4 = list2.pop()
print(list2)
print("v4:",v4)
v5 = list2.pop()
print(list2)
print("v5:",v5)
```

### Del
+ list 可透過 del 方式移除指定位置資料。
+ del list[ ] 必須指定範圍：
  + del list[n] 代表移除第 n 位索引值資料。
  + del list[:m] 代表移除第 m 位索引值之前 ( 不包含 m ) 資料。
  + del list[n:m] 代表移除第 n 到 m 位索引值之前 ( 不包含 m ) 資料。

```python
list2 = ["A","B","C","D","E"]
del list2[2]
print(list2)
del list2[1:3]
print(list2)
```

## isinstance

isinstance(x,int) 代表判斷x是否為int型態

list若要進行排序得先挑選出相同型態資料。

請使用isinstance(資料,資料型態)方式於list中撈出特定資料。

list2a=[x for x in list2 if isinstance(x,int)]

這樣的寫法的目的

避免程式流程受到外面的干擾

流程單純化，list的資料變化就在list區域內進行

促進語法的可讀性提高!

```python
list1 = ['x',1,'y',2,'z',3]
numList = [v for  v in list1 if isinstance(v,int) ]
strList = [v for v in list1 if isinstance(v,str)]
print(numList)
print(strList)
```
