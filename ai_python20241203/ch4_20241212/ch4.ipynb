````python
h = int(input("請輸入身高"))
w = float(input("請輸入體重"))
bmi = w / (h / 100) ** 2
if bmi < 18.5:
    print("太輕")
elif bmi < 25:
    print("正常")
elif bmi < 30:
    print("過重")
else:
    print("胖胖")
````

````python
for i in range(4,8):
    print(i,end = " ")
print()    
print("=================")    
for i in range(-4,-8,-1):
    print(i, end = "")
print()    
print("=================")   
for i in range(-8,-4):
    print(i,end = " ")
````

````python
for i in range(1,4):
    for x in range(3,8):
        print(i,"_",x,sep="",end = " ")
    print()    
````

````python
n = 'y'
while n == 'y':
    a = int(input('請輸入第一個數字'))
    b = int(input('請輸入第二個數字'))
    ans = a * b
    print(a,"*",b,"=",ans)
    n = input("是否繼續y/n")
    n = n.lower()#轉小寫
````

## break 代表不回頭，跳出迴圈。
## continue 代表繼續執行下一個迴圈

````python
for i in range(1,11):
    if i == 5:
        break
    print(i,end = " ")
````

````python
for i in range(1,11):
    if i == 5 or i == 7:
        continue
    print(i,end=" ")    
````

````python
print("請輸入大於80的整數")#持續數輸入直到輸入數值是大於80為止
while True:
    a = int( input("請輸入數字"))
    if a > 80:
        print("a=",a)
        break
print("離開迴圈")        
````
